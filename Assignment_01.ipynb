Python keywords are reserved words that have special meanings within the language. They are essential for defining the syntax and structure of Python programs. Keywords are used to control the flow of the program, define variables, handle errors, create functions, and more. Because of their predefined role in the language, keywords cannot be used as variable names or identifiers.

How Python Keywords Are Used in Programs
Let's explore the role of Python keywords with examples:

**a. Control Flow Keywords**

These keywords manage the flow of execution in a program.

**if, elif, else:-** Used for conditional branching.

**Example:-**

       print("You are an adult.")

       print("You are a teenager.")

       print("You are a child.")

**Mutable Objects**

Mutable objects are those that can be modified after they are created. This means that the object's contents can be changed in place, without changing the object's identity.

**Examples of Mutable Objects:-**

**Lists**

Lists can be modified by adding, removing, or changing elements.

**Example:-**

    my_list = [1, 2, 3]

    my_list[0] = 10  # Changing the first element
    print(my_list)  # Outputs: [10, 2, 3]

    my_list.append(4)  # Adding an element
    print(my_list)  # Outputs: [10, 2, 3, 4]

**Dictionaries**

Dictionaries allow modification by adding, removing, or updating key-value pairs.

**Example:-**

    my_dict = {'a': 1, 'b': 2}
    my_dict['a'] = 10  # Updating the value of 'a'
    print(my_dict)  # Outputs: {'a': 10, 'b': 2}

    my_dict['c'] = 3  # Adding a new key-value pair
    print(my_dict)  # Outputs: {'a': 10, 'b': 2, 'c': 3}

**Sets**

Sets can have elements added or removed.

**Example:-**

    my_set = {1, 2, 3}
    my_set.add(4)  # Adding an element
    print(my_set)  # Outputs: {1, 2, 3, 4}

    my_set.remove(2)  # Removing an element
    print(my_set)  # Outputs: {1, 3, 4}

**Immutable Objects**

Immutable objects are those that cannot be modified after they are created. Any attempt to alter the contents of an immutable object will result in the creation of a new object with the modified value, rather than changing the original object.

**Examples of Immutable Objects:-**

**Strings**

Strings cannot be changed in place; operations that modify a string result in a new string.

**Example:-**

    my_string = "hello"
    new_string = my_string.replace('h', 'j')
    print(my_string)  # Outputs: hello (original string remains unchanged)
    print(new_string)  # Outputs: jello (new string created)

**Tuples**

Tuples are like lists but are immutable; their elements cannot be changed after creation.

**Example:-**

    my_tuple = (1, 2, 3)
    # my_tuple[0] = 10  
    # This would raise a TypeError

    new_tuple = my_tuple + (4,)
    print(my_tuple)  # Outputs: (1, 2, 3) (original tuple remains unchanged)
    print(new_tuple)  # Outputs: (1, 2, 3, 4) (new tuple created)

**Integers, Floats, and Booleans**

These basic data types are immutable; any operation that changes their value results in a new object.

**Example:-**

    x = 5
    y = x + 1
    print(x)  # Outputs: 5 (original integer remains unchanged)
    print(y)  # Outputs: 6 (new integer created)

   **5. Explain the concept of Type Casting in Python with examples.**

Type casting in Python refers to the process of converting one data type into another. This can be done explicitly by the programmer using built-in functions or implicitly by Python itself when performing certain operations. Here's a detailed look at type casting in Python:



**a. Explicit Type Casting**

Explicit type casting involves converting data from one type to another using Python's built-in functions. This is controlled directly by the programmer.

**Common Type Casting Functions:-**

* **int():-** Converts a value to an integer.

**Example:-**

    num_str = "10"
    num_int = int(num_str)  
        # Converts the string "10" to the integer 10
* **float():-** Converts a value to a floating-point number.

**Example:-**

    num_int = 10
    num_float = float(num_int)  
        # Converts the integer 10 to the float 10.0

* **str():-** Converts a value to a string.

**Example:-**

    num_int = 10
    num_str = str(num_int)  
        # Converts the integer 10 to the string "10"

* **list():-** Converts an iterable (like a tuple or string) to a list.

**Example:-**

    tuple_data = (1, 2, 3)
    list_data = list(tuple_data)  
        # Converts the tuple (1, 2, 3) to the list [1, 2, 3]

* **tuple():-** Converts an iterable (like a list) to a tuple.

**Example:-**

    list_data = [1, 2, 3]
    tuple_data = tuple(list_data)  
        # Converts the list [1, 2, 3] to the tuple (1, 2, 3)

* **set():-** Converts an iterable to a set.

**Example:-**

    list_data = [1, 2, 2, 3]
    set_data = set(list_data)  
        # Converts the list [1, 2, 2, 3] to the set {1, 2, 3}

* **dict():-** Converts a sequence of key-value pairs into a dictionary.

**Example:-**

    tuple_data = (("a", 1), ("b", 2))
    dict_data = dict(tuple_data)  
        # Converts the tuple to the dictionary {"a": 1, "b": 2}

**More Examples of Explicit Type Casting:-**

* **String to Integer:-**


    s = "123"
    n = int(s)  
        # Converts "123" to 123

* **Integer to String:-**


    n = 123
    s = str(n)  
        # Converts 123 to "123"

* **Float to Integer:-**

    f = 123.45
    n = int(f)  
        # Converts 123.45 to 123 (truncates the decimal part)

**Integer to Float:-**

    n = 123
    f = float(n)  
        # Converts 123 to 123.0

* **List to String (using join):-**

    lst = ['a', 'b', 'c']
    s = ''.join(lst)  
        # Converts ['a', 'b', 'c'] to "abc"

**b. Implicit Type Casting**

Python performs implicit type casting when necessary, usually during operations between different types. This is automatic and done behind the scenes to avoid type errors.

**Examples of Implicit Type Casting:-**

* **Integer to Float:-**

When an integer is involved in an operation with a float, Python automatically converts the integer to a float to avoid data loss.

    result = 10 + 5.0  
        # 10 is converted to 10.0, so result is 15.0

* **Boolean to Integer:-**

Python considers True as 1 and False as 0 during arithmetic operations.


    result = True + 5
        # True is converted to 1, so result is 6

* **String Concatenation:-**

Python will not automatically convert numbers to strings when concatenating with strings; explicit conversion is required.

    result = "The number is " + str(10)  
        # Converts 10 to "10" and concatenates

**c. Type Casting in Complex Scenarios**

**Casting Within Data Structures:-**

*  **Casting Elements in a List:**

You can use type casting within list comprehensions to convert the type of each element.

    str_list = ["1", "2", "3"]
    int_list = [int(i) for i in str_list]  
        # Converts each element to an integer

* **Nested Data Structure Casting:-**

If you have a list of tuples, you might need to cast elements within those tuples.

    data = [("1", "2"), ("3", "4")]
    int_data = [(int(a), int(b)) for a, b in data]

*  **Casting Return Values:-**

Sometimes, you need to cast the return value of a function to fit the required type.

    def get_number():-
        #return "42"

**d. Limitations and Considerations**

**Invalid Casts:-**

Not all type casts are valid. For instance, trying to convert a string that doesn't represent a number into an integer will result in a ValueError.

    s = "abc"
    n = int(s)

When casting from a float to an integer, the decimal part is truncated, which can lead to a loss of precision.


    f = 123.99

Sometimes, casting is necessary to ensure the consistency of data types, especially when combining different data sources or performing arithmetic operations.

Type casting is an essential concept in Python, enabling the flexible handling of different data types and ensuring that operations perform as expected. Understanding when and how to use type casting is crucial for writing robust and error-free Python code.

**6. How conditional statement work in Python ? Illustrate with example**

Conditional statements in Python allow you to execute different blocks of code based on certain conditions. These statements evaluate expressions and, depending on whether the expression is True or False, the corresponding code block is executed. Python primarily uses if, elif, and else keywords to implement conditional logic.

**Structure of Conditional Statements**

**if Statement:-** The if statement checks whether a condition is True. If it is, the code block following the if statement is executed.

**elif Statement:-** The elif (short for "else if") statement checks another condition if the previous if condition is False. You can have multiple elif statements.

**else Statement:-** The else statement executes a code block if all preceding if and elif conditions are False.

**Syntax**

    if condition1:
        # Code to execute if condition1 is True
    elif condition2:
        # Code to execute if condition1 is False and condition2 is True
    else:
        # Code to execute if both condition1 and condition2 are False

**Example 1: Basic Conditional Statement**

    age = 20

    if age >= 18:
        print("You are an adult.")
    else:
        print("You are a minor.")
**Explanation:-**

The condition age >= 18 is checked.
Since age is 20, the condition is True, so the code block under the if statement is executed.
Output: You are an adult.

**Example 2:- Using elif for Multiple Conditions**

    marks = 75
    if marks >= 90:
        print("Grade: A")
    elif marks >= 80:
        print("Grade: B")
    elif marks >= 70:
        print("Grade: C")
    else:
        print("Grade: D")
**Explanation:-**

The program checks the conditions in order:-

    marks >= 90: False
    marks >= 80: False
    marks >= 70: True
    The block under the elif marks >= 70: is executed.
    Output: Grade: C

**Example 3:- Nested Conditional Statements**

    number = -10
    if number >= 0:
      if number == 0:
        print("Number is zero.")
      else:
        print("Number is positive.")
    else:
      print("Number is negative.")
**Explanation:-**

The outer if checks if number >= 0.
Since number is -10, the outer else block is executed.
Output: Number is negative.

**Example 4:- Conditional Statements with Logical Operators**

    temp = 30
    if temp > 0 and temp < 20:
        print("It's cold.")
    elif temp >= 20 and temp <= 30:
        print("It's warm.")
    else:
        print("It's hot.")
**Explanation:-**

The condition temp > 0 and temp < 20 is False.
The condition temp >= 20 and temp <= 30 is True.
The block under the elif is executed.
Output: It's warm.

**Example 5:- if with Inline Condition**

You can write simple conditional statements in a single line using a shorthand syntax.

    age = 18
    status = "Adult" if age >= 18 else "Minor"
    print(status)
**Explanation:-**

The condition age >= 18 is checked.
Since it's True, the value "Adult" is assigned to status.
Output: Adult

**Summary**

    if: Checks the first condition.
    elif: Checks subsequent conditions if the previous ones were False.
    else: Executes a block of code if none of the preceding conditions were True.
    Nested Conditions: You can place if, elif, and else statements inside other if, elif, and else statements for more complex logic.
    Logical Operators: and, or, and not can combine multiple conditions within a single if or elif statement.


**7. Describe the different types of loop in python  and thier use cases with example.**

Python provides several types of loops that allow you to repeat blocks of code. The primary loop structures in Python are for loops and while loops, each with its own use cases. Additionally, there are control statements like break, continue, and pass that modify the behavior of loops.

**a. for Loop**
The for loop in Python is used to iterate over a sequence (like a list, tuple, string, or range) or other iterable objects. The loop runs once for each item in the sequence, executing the block of code inside the loop.

**Use Cases:-**
Iterating over a sequence.
Accessing elements of a collection one by one.
Executing a block of code a specific number of times.

**Example 1:-** Iterating Over a List
    fruits = ["apple", "banana", "cherry"]
    for fruit in fruits:
        print(fruit)
**Output:-**

    apple
    banana
    cherry
**Explanation:-** The loop iterates over the fruits list, printing each fruit in the list.

**Example 2:-** Using range() with for Loop

    for i in range(5):
        print(i)
**Output:-**

    0
    1
    2
    3
    4
**Explanation:-** The loop runs 5 times, with i taking values from 0 to 4 (inclusive), provided by range(5).

**b. while Loop**

The while loop in Python repeatedly executes a block of code as long as a specified condition is True. The condition is evaluated before each iteration of the loop.

**Use Cases:-**
Repeating actions until a condition is no longer true.
Waiting for a condition to change inside the loop.
When the number of iterations is not known beforehand.

**Example 1:-** Basic while Loop
    count = 0

    while count < 5:
        print(count)
        count += 1
**Output:-**

    0
    1
    2
    3
    4
**Explanation:-** The loop continues to run as long as count is less than 5. In each iteration, count is incremented by 1.

**Example 2:-** while Loop with User Input

    password = ""
    while password != "12345":
        password = input("Enter the password: ")
    print("Access granted.")

**Explanation:-** The loop keeps asking the user to enter the correct password until the user enters "12345". Once the correct password is entered, the loop exits, and "Access granted." is printed.

**c. Control Statements in Loops**

Control statements are used to modify the behavior of loops.

**break:-** Exits the loop prematurely, regardless of the loop's condition.

**Example:-**

    for i in range(10):
        if i == 5:
            break
        print(i)
**Output:-**

    0
    1
    2
    3
    4
**Explanation:-** The loop exits when i equals 5.

continue: Skips the current iteration and proceeds to the next iteration of the loop.

**Example:-**
    for i in range(5):
        if i == 2:
            continue
        print(i)

**Output:-**

    0
    1
    3
    4
**Explanation:-** When i equals 2, the continue statement skips the rest of the loop's body for that iteration.

**pass:-** Acts as a placeholder and does nothing. It's often used in loops or functions that you plan to implement later.

**Example:-**

    for i in range(5):
        if i == 2:
            pass  # No operation is performed when i is 2
        print(i)
**d. else Clause with Loops**

Python allows the use of an else clause with both for and while loops. The else block is executed when the loop terminates naturally (i.e., not via a break statement).

**Use Cases:-**

Running code after a loop completes all its iterations.
Checking if a loop was exited prematurely with a break.

**Example:-** else with for Loop

    for i in range(5):
        print(i)
    else:
        print("Loop completed successfully.")
**Output:-**

    0
    1
    2
    3
    4
    Loop completed successfully.
**Explanation:-** The else block runs after the loop finishes iterating over the range.

**Example:-** else with while Loop

    count = 0
    while count < 5:
        print(count)
        count += 1
    else:
        print("Loop completed successfully.")
**Output:-**

    0
    1
    2
    3
    4
    Loop completed successfully.
**Explanation:-** The else block runs after the while loop's condition becomes False.
